# Handling Missing Values using Scikit-Learn

```python
from sklearn.preprocessing import Imputer
```

```python
imputer = Imputer(missing_values = 'NaN', strategy = "mean", axis = 0)
```

```python
data
```

```python
imputer = imputer.fit(data.iloc[:,1:3])
```

```python
data.iloc[:,1:3] = imputer.transform(data.iloc[:,1:3])
```

```python
data
```

# imputer will work only for numbers not for string values so for string we need to use mood

```python
col = ['Purchased']
data[col] = data[col].fillna(data.mode().iloc[0])
```

```python
data
```

# Mean Imputation :gives mean value for missing data

```python
data.fillna(data.mean())
```

# median imputation : Gives median value for missing values in given data

```python
data.fillna(data.median())
```
