# Print no. of matches won by each team based on year wise (each year as one column)


```python
import pandas as pd
def matchwon(filename):
    df=pd.read_csv(filename)
    a=df.groupby(['Season_Year','match_winner']).match_winner.count().unstack(level=0).fillna(0)
    return a
matchwon('datafiles/Match.csv')
```

```python
import pandas as pd
import numpy as np
def manofmatch(filename):
    df=pd.read_csv(filename)
    li=[]
    counts=[]
    d={}
    for row in df.values:
        if(row[7]=='Hyderabad') or (row[7]=='Hyderabad (Deccan)'):
            li.append (row[14])
    uniq,counts=np.unique(li,return_counts=True)
    d={'man_of_match':uniq,'count':counts}
    dff=pd.DataFrame(d)
    return dff
manofmatch('datafiles/Match.csv')
            
        
        
    
```

```python
import pandas as pd
import numpy as np
def matchwonandloss(filename):
    team=[]
    uniqueteam=[]
    df=pd.read_csv(filename)
    for row in df.values:
        team.append (row[10])
        uniqueteam=np.unique(team)
    winlist=[]
    losslist=[]
    for i in uniqueteam:
        wc=0
        lc=0
        for row in df.values:
            if i==row[9] and i==row[10]:
                wc=wc+1
            elif i==row[2] or i==row[3]:
                lc=lc+1
        winlist.append (wc)
        losslist.append(lc)
    d={'team':uniqueteam,'win':winlist,'loss':losslist}
    dff=pd.DataFrame(d)
    return dff
matchwonandloss('datafiles/Match.csv')
        
         
```
